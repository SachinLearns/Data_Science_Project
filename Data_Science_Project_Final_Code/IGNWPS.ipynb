```python
# Drop the data with 'fin_year' as '2024-2025' as it is not in our scope
data6 = data6[data6['fin_year'] != '2024-2025']
data6.reset_index(inplace=True, drop=True)
data6
```

```python
# Explicitly define the Zone column as an object type
data6['Zone'] = pd.Series(dtype='object')

# Assign the zones to the respective states
for index, row in data6.iterrows():
    state = row['state_name'].strip().lower()
    zone_assigned = False

    for zone, divisions in india_zones.items():

        if any(state == place.lower() for place in divisions.get("States", []) + divisions.get("Union Territories", [])):
            data6.at[index, 'Zone'] = zone
            zone_assigned = True
            break

    if not zone_assigned:
        data6.at[index, 'Zone'] = 'Unassigned'

data6 = data6[['Zone'] + [col for col in data6.columns if col != 'Zone']]
data6
```

```python
# Combine the zones into a single row and add the corresponding columns
data6_grouped = data6.groupby('Zone').sum().reset_index()
# Drop the state_name column
data6_grouped.drop(columns='state_name', inplace=True)
data6_grouped
```

# Hypothesis Testing

*   **Null Hypothesis**: More than 50% of the beneficiaries in each zone are general category candidates
*   **None Hypothesis**: More than 50% of the beneficiaries in each zone are not the general category candidates

```python
import numpy as np
from scipy.stats import norm
import pandas as pd

# Sample 50 rows (or less if not available) from each zone without altering the grouping columns
sampled_data6 = data6.groupby('Zone', group_keys=False).apply(lambda x: x.sample(min(len(x), 50), random_state=1)).reset_index(drop=True)

# Summarize the sampled data by zone
zone_sample_summary = sampled_data6.groupby('Zone').agg({
    'total_beneficiaries': 'sum',
    'total_gen': 'sum'
}).reset_index()

# Hypothesized proportion for null hypothesis (50% of beneficiaries are general category)
p_hypothesis = 0.50  # 50%
alpha = 0.05  # significance level for one-tailed test

# Perform Z-test for proportions
z_test_results = []
for _, row in zone_sample_summary.iterrows():
    zone = row['Zone']
    total = row['total_beneficiaries']
    general = row['total_gen']  # General category beneficiaries
    
    # Observed proportion
    p_observed = general / total
    
    # Standard error
    se = np.sqrt(p_hypothesis * (1 - p_hypothesis) / total)
    
    # Z-score (one-tailed test)
    z_score = (p_observed - p_hypothesis) / se
    
    # p-value for one-tailed test (right-tail since we are testing if it's greater than 50%)
    # Update p-value for up to 5 decimal places
    p_value = np.round(1 - norm.cdf(z_score), 5)
    
    # Determine result based on p-value
    result = "Reject H0" if p_value > alpha else "Fail to Reject H0"
    
    # Store results
    z_test_results.append({
        'Zone': zone,
        'Total Beneficiaries': total,
        'General Beneficiaries': general,
        'Observed Proportion': p_observed,
        'Hypothesis Proportion': p_hypothesis,
        'Z-Score': z_score,
        'P-Value': p_value,
        'Result': result
    })

# Convert results to DataFrame and output results
z_test_results_df = pd.DataFrame(z_test_results)
z_test_results_df
```
