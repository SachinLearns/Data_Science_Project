It seems that the PMGSY scheme is not a single scheme but a collection of schemes. It contains five different schemes as listed under the column 'PMGSY_SCHEME'.

On some internet search, it was found that these are different legs of the broader PMGSY scheme, launched either as a phase-wise scheme or with a focus in hand. The five schemes are:

1. **PMGSY-I (Pradhan Mantri Gram Sadak Yojana - Phase I):**
   Focused on providing all-weather road connectivity to unconnected rural habitations, mainly targeting villages with populations over 500 in plains and 250 in hilly, tribal, or desert areas.

2. **PMGSY-II (Pradhan Mantri Gram Sadak Yojana - Phase II):**
   Launched to upgrade and consolidate the rural road network by improving existing roads that connect rural areas to key facilities like markets, schools, and health centers.

3. **PMGSY-III (Pradhan Mantri Gram Sadak Yojana - Phase III):**
   Aims to further enhance rural road connectivity by constructing roads that improve access to agricultural markets, higher education institutions, and healthcare services, promoting rural economic growth.

4. **RCPLWEA (Road Connectivity Project for Left Wing Extremism Affected Areas):**
   Designed to improve road infrastructure in regions affected by Left-Wing Extremism, with the goal of enhancing mobility, security, and development in these insurgency-hit areas.

5. **PM-JANMAN (Pradhan Mantri Janjatiya Adivasi Nyay Mahabhiyan):**
   Focused on providing justice, development, and infrastructure, especially road connectivity, to tribal and Adivasi communities, addressing socio-economic inequalities and promoting their welfare.


Thus, we will consider each of these schemes as a separate scheme and analyze them individually.

```python
# Dividing the dataset into the 5 schemes (['PMGSY-II' 'PMGSY-I' 'PMGSY-III' 'RCPLWEA' 'PM-JANMAN'])

scheme1 = data1[data1['PMGSY_SCHEME'] == 'PMGSY-I']
scheme2 = data1[data1['PMGSY_SCHEME'] == 'PMGSY-II']
scheme3 = data1[data1['PMGSY_SCHEME'] == 'PMGSY-III']
scheme4 = data1[data1['PMGSY_SCHEME'] == 'RCPLWEA']
scheme5 = data1[data1['PMGSY_SCHEME'] == 'PM-JANMAN']

# Drop the 'PMGSY_SCHEME' column from each scheme dataset
scheme1 = scheme1.drop(columns=['PMGSY_SCHEME'])
scheme2 = scheme2.drop(columns=['PMGSY_SCHEME'])
scheme3 = scheme3.drop(columns=['PMGSY_SCHEME'])
scheme4 = scheme4.drop(columns=['PMGSY_SCHEME'])
scheme5 = scheme5.drop(columns=['PMGSY_SCHEME'])

# Since we are not concerned about the district-level information of the dataset, we can drop the 'DISTRICT_NAME' column
scheme1 = scheme1.drop(columns=['DISTRICT_NAME'])
scheme2 = scheme2.drop(columns=['DISTRICT_NAME'])
scheme3 = scheme3.drop(columns=['DISTRICT_NAME'])
scheme4 = scheme4.drop(columns=['DISTRICT_NAME'])
scheme5 = scheme5.drop(columns=['DISTRICT_NAME'])
```

```python
# State-wise data grouping

scheme1_statewise = scheme1.groupby('STATE_NAME').sum()
scheme1_statewise = scheme1_statewise.reset_index()
# scheme1_statewise

scheme2_statewise = scheme2.groupby('STATE_NAME').sum()
scheme2_statewise = scheme2_statewise.reset_index()

scheme3_statewise = scheme3.groupby('STATE_NAME').sum()
scheme3_statewise = scheme3_statewise.reset_index()

scheme4_statewise = scheme4.groupby('STATE_NAME').sum()
scheme4_statewise = scheme4_statewise.reset_index()

scheme5_statewise = scheme5.groupby('STATE_NAME').sum()
scheme5_statewise = scheme5_statewise.reset_index()
scheme5_statewise
```

```python
# Assigning the 'ZONE' to each state

# Adding the 'ZONE' column to each scheme dataset and assigning the zone name to each state
for zone, divisions in india_zones.items():
    for division, places in divisions.items():
        for place in places:
            scheme1_statewise.loc[scheme1_statewise['STATE_NAME'] == place, 'ZONE'] = zone
            scheme2_statewise.loc[scheme2_statewise['STATE_NAME'] == place, 'ZONE'] = zone
            scheme3_statewise.loc[scheme3_statewise['STATE_NAME'] == place, 'ZONE'] = zone
            scheme4_statewise.loc[scheme4_statewise['STATE_NAME'] == place, 'ZONE'] = zone
            scheme5_statewise.loc[scheme5_statewise['STATE_NAME'] == place, 'ZONE'] = zone


# shift the 'ZONE' column to the first position
cols = scheme1_statewise.columns.tolist()
cols = cols[-1:] + cols[:-1]
scheme1_statewise = scheme1_statewise[cols]

cols = scheme2_statewise.columns.tolist()
cols = cols[-1:] + cols[:-1]
scheme2_statewise = scheme2_statewise[cols]

cols = scheme3_statewise.columns.tolist()
cols = cols[-1:] + cols[:-1]
scheme3_statewise = scheme3_statewise[cols]

cols = scheme4_statewise.columns.tolist()
cols = cols[-1:] + cols[:-1]
scheme4_statewise = scheme4_statewise[cols]

cols = scheme5_statewise.columns.tolist()
cols = cols[-1:] + cols[:-1]
scheme5_statewise = scheme5_statewise[cols]

scheme1_statewise
```

```python
# Zone-wise Grouping

scheme1_zonewise = scheme1_statewise.groupby('ZONE').sum()
scheme1_zonewise = scheme1_zonewise.reset_index()
# scheme1_zonewise

scheme2_zonewise = scheme2_statewise.groupby('ZONE').sum()
scheme2_zonewise = scheme2_zonewise.reset_index()

scheme3_zonewise = scheme3_statewise.groupby('ZONE').sum()
scheme3_zonewise = scheme3_zonewise.reset_index()

scheme4_zonewise = scheme4_statewise.groupby('ZONE').sum()
scheme4_zonewise = scheme4_zonewise.reset_index()

scheme5_zonewise = scheme5_statewise.groupby('ZONE').sum()
scheme5_zonewise = scheme5_zonewise.reset_index()
scheme5_zonewise
```

The 'NO_OF_ROAD_WORKS_BALANCE' & 'NO_OF_BRIDGES_BALANCE' give no new information when the data about the completed works is already available. Hence, we will drop these columns.

Similarly, 'LENGTH_OF_ROAD_WORK_BALANCE_KM' gives no new information when the data about the completed works is already available. Hence, we will drop this column.

```python
# Drop 'NO_OF_ROAD_WORKS_BALANCE' & 'NO_OF_BRIDGES_BALANCE'

scheme1_zonewise = scheme1_zonewise.drop(columns=['NO_OF_ROAD_WORKS_BALANCE', 'NO_OF_BRIDGES_BALANCE', 'LENGTH_OF_ROAD_WORK_BALANCE_KM'], axis=1)
scheme1_zonewise.columns
```

Since the number of roads and bridges do not have a distinction with respect to the problem in hand, we can combine them to get the total number of roads and bridges constructed under the scheme.

```python
# Combining 'NO_OF_ROAD_WORK_SANCTIONED' and 'NO_OF_BRIDGES_SANCTIONED'
scheme1_zonewise['NO_OF_ROAD_WORK_AND_BRIDGES_SANCTIONED'] = scheme1_zonewise['NO_OF_ROAD_WORK_SANCTIONED'] + scheme1_zonewise['NO_OF_BRIDGES_SANCTIONED']
scheme1_zonewise = scheme1_zonewise.drop(columns=['NO_OF_ROAD_WORK_SANCTIONED', 'NO_OF_BRIDGES_SANCTIONED'])

# Similarly, combining 'NO_OF_ROAD_WORKS_COMPLETED' & 'NO_OF_BRIDGES_COMPLETED'
scheme1_zonewise['NO_OF_ROAD_WORKS_AND_BRIDGES_COMPLETED'] = scheme1_zonewise['NO_OF_ROAD_WORKS_COMPLETED'] + scheme1_zonewise['NO_OF_BRIDGES_COMPLETED']
scheme1_zonewise = scheme1_zonewise.drop(columns=['NO_OF_ROAD_WORKS_COMPLETED', 'NO_OF_BRIDGES_COMPLETED'])

scheme1_zonewise.columns
```

### Length to Expenditure Ratio

```python
plt.figure(figsize=(10, 10))

# Create a scatter plot with a regression line (without confidence interval)
sns.regplot(x='LENGTH_OF_ROAD_WORK_SANCTIONED_KM', 
            y='COST_OF_WORKS_SANCTIONED_LAKHS', 
            data=scheme1_zonewise, 
            scatter_kws={'s': 50}, # s: size of the points in the scatter plot
            ci = 95, # 95% confidence interval 
            line_kws={"color": "red"})

# Overlay the scatter plot with hue based on the 'ZONE'
markers = ['o', 's', 'D', '^', 'v', 'X']  # Circle, Square, Diamond, Triangle up, Triangle down, X
sns.scatterplot(x='LENGTH_OF_ROAD_WORK_SANCTIONED_KM', 
                y='COST_OF_WORKS_SANCTIONED_LAKHS', 
                data=scheme1_zonewise,
                hue='ZONE', 
                style='ZONE',  # Differentiating by markers
                markers=markers,
                palette=colors, # Add the color same as the pie chart
                s=300) # s: size of the points in the scatter plot

plt.legend(loc='best', title='ZONE', ncol=2, markerscale=1.5, borderpad=1.5, labelspacing=2.5, handletextpad=1)
plt.title('Comparison of LENGTH_OF_ROAD_WORK_SANCTIONED_KM and COST_OF_WORKS_SANCTIONED_LAKHS with Trend Line')
plt.show()
```

*   The red **regression line** shows the general relationship between the length of roadwork sanctioned (x-axis) and the cost of works sanctioned (y-axis).
*   The **upward slope** of the regression line indicates a positive correlation, meaning that as the length of roadwork sanctioned increases, the cost sanctioned also increases, which is expected and aligns with intuition.
*   However, **some points significantly deviate** from the regression line, such as the data points for the North-Eastern Zone (green rhombus), Central Zone (red circle), and Eastern Zone (blue square). These deviations suggest that in these zones, the cost of works sanctioned does not strictly follow the expected proportionality.

### Completion Rate

```python
# Zone-wise percentage of roads completed out of the total roads sanctioned

scheme1_zonewise['PERCENTAGE_OF_ROAD_WORKS_COMPLETED'] = (scheme1_zonewise['LENGTH_OF_ROAD_WORK_COMPLETED_KM'] / scheme1_zonewise['LENGTH_OF_ROAD_WORK_SANCTIONED_KM']) * 100
scheme1_zonewise['PERCENTAGE_OF_ROAD_WORKS_COMPLETED'] = scheme1_zonewise['PERCENTAGE_OF_ROAD_WORKS_COMPLETED'].round(2)

scheme1_zonewise['PERCENTAGE_OF_ROAD_WORKS_COMPLETED']

# Visualize the bar chart
plt.figure(figsize=(10, 10))
bars = plt.bar(scheme1_zonewise['ZONE'], 
               scheme1_zonewise['PERCENTAGE_OF_ROAD_WORKS_COMPLETED'], 
               color=colors)

# Add percentage labels at the top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval}%', va='bottom', ha='center', fontsize=12)

plt.xlabel('Zones', fontsize=14)
plt.ylabel('Percentage of Roads Completed', fontsize=14)
plt.title('Percentage of Roads Completed in each Zone in PMGSY-I Scheme', fontsize=16)
plt.tight_layout()
plt.show()

# Drop the 'PERCENTAGE_OF_ROAD_WORKS_COMPLETED' column
scheme1_zonewise = scheme1_zonewise.drop(columns=['PERCENTAGE_OF_ROAD_WORKS_COMPLETED'])
```

This plot shows the completion rate of roadworks across different zones. The completion rate is calculated as the ratio of the number of completed roadworks to the total number of sanctioned roadworks. The completion rate is an important metric to assess the efficiency and effectiveness of the scheme's implementation.

Here, we observe that the completion rates of all the zones are relatively close to each other, ranging from 0.95 to 0.97. This indicates that the completion rates are consistent across the zones, suggesting a similar level of efficiency in implementing the PMGSY-I scheme.
